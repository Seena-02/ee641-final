```python
import os
from PIL import Image
import imagehash
from collections import defaultdict

# ---- CONFIG ----
image_folder = "data_7x7/grids/both"   # << CHANGE THIS
valid_ext = {".jpg", ".jpeg", ".png", ".bmp", ".gif", ".tiff"}

# ---- STEP 1: Compute image hashes ----
hash_dict = defaultdict(list)

for filename in os.listdir(image_folder):
    ext = os.path.splitext(filename)[1].lower()
    if ext not in valid_ext:
        continue  # skip non-images

    filepath = os.path.join(image_folder, filename)

    try:
        img = Image.open(filepath)
        img_hash = imagehash.phash(img)  # perceptual hash
        hash_dict[str(img_hash)].append(filename)
    except Exception as e:
        print(f"Error reading {filename}: {e}")

# ---- STEP 2: Print duplicates ----
duplicates = {h: files for h, files in hash_dict.items() if len(files) > 1}

if duplicates:
    print("🔍 Found duplicate images (visually identical):\n")
    for h, files in duplicates.items():
        print(f"Hash: {h}")
        for f in files:
            print(f"  - {f}")
        print()
else:
    print("✅ No duplicate images found.")
```

```python
import os
from PIL import Image
import imagehash
from collections import defaultdict

# ---- CONFIG ----
image_folder = "data_4x4/grids/both"   # << CHANGE THIS
valid_ext = {".jpg", ".jpeg", ".png", ".bmp", ".gif", ".tiff"}

# ---- STEP 1: Compute hashes ----
hash_dict = defaultdict(list)

for filename in os.listdir(image_folder):
    ext = os.path.splitext(filename)[1].lower()
    if ext not in valid_ext:
        continue

    filepath = os.path.join(image_folder, filename)

    try:
        img_hash = imagehash.phash(Image.open(filepath))
        hash_dict[str(img_hash)].append(filename)
    except Exception as e:
        print(f"Error reading {filename}: {e}")

# ---- STEP 2: Count duplicates ----
duplicate_groups = [files for files in hash_dict.values() if len(files) > 1]

total_duplicate_images = sum(len(group) for group in duplicate_groups)

print("📸 Total groups of duplicates:", len(duplicate_groups))
print("🔢 Total duplicate images:", total_duplicate_images)
print("\nGroups (for verification):")

for group in duplicate_groups:
    print(group)
```
