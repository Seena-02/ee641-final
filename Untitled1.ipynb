```python
import json
import torch
from torch.utils.data import Dataset, DataLoader
from PIL import Image
from transformers import GPT2TokenizerFast, ViTImageProcessor, VisionEncoderDecoderModel

# ---------------------------
# 1️⃣ Setup MPS device for Mac
# ---------------------------
device = torch.device("mps") if torch.backends.mps.is_available() else torch.device("cpu")
print(f"Using device: {device}")

# ---------------------------
# 2️⃣ Dataset class with attention mask
# ---------------------------
class JSONImageCaptionDataset(Dataset):
    def __init__(self, json_path, image_root, image_processor, tokenizer, max_length=128):
        with open(json_path, 'r') as f:
            self.data = json.load(f)
        self.image_root = image_root
        self.image_processor = image_processor
        self.tokenizer = tokenizer
        self.max_length = max_length

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        item = self.data[idx]
        image_path = f"{self.image_root}/{item['image']}"
        image = Image.open(image_path).convert("RGB")
        pixel_values = self.image_processor(image, return_tensors="pt").pixel_values.squeeze(0)
        
        caption_str = " ".join(item["sequence"])
        
        encoding = self.tokenizer(
            caption_str,
            padding="max_length",
            truncation=True,
            max_length=self.max_length,
            return_tensors="pt",
            return_attention_mask=True
        )
        labels = encoding.input_ids.squeeze(0)
        attention_mask = encoding.attention_mask.squeeze(0)
        labels[labels == self.tokenizer.pad_token_id] = -100
        
        return {
            "pixel_values": pixel_values,
            "labels": labels,
            "attention_mask": attention_mask
        }

# ---------------------------
# 3️⃣ Load model, tokenizer, processor
# ---------------------------
model_name = "nlpconnect/vit-gpt2-image-captioning"
model = VisionEncoderDecoderModel.from_pretrained(model_name)

tokenizer = GPT2TokenizerFast.from_pretrained(model_name)
tokenizer.pad_token = tokenizer.eos_token

image_processor = ViTImageProcessor.from_pretrained(model_name)

model.config.pad_token_id = tokenizer.pad_token_id
model.config.decoder_start_token_id = tokenizer.bos_token_id
model.config.vocab_size = tokenizer.vocab_size
model.to(device)

# ---------------------------
# 4️⃣ Dataset + DataLoader
# ---------------------------
json_path = "data/grid_sequences.json"
image_root = "data"

dataset = JSONImageCaptionDataset(json_path, image_root, image_processor, tokenizer, max_length=128)
dataloader = DataLoader(dataset, batch_size=4, shuffle=True)

# ---------------------------
# 5️⃣ Training loop
# ---------------------------
optimizer = torch.optim.AdamW(model.parameters(), lr=5e-5)
num_epochs = 5
grad_accum_steps = 2

model.train()
for epoch in range(num_epochs):
    running_loss = 0.0
    optimizer.zero_grad()
    
    for step, batch in enumerate(dataloader):
        pixel_values = batch["pixel_values"].to(device)
        labels = batch["labels"].to(device)
        decoder_attention_mask = batch["attention_mask"].to(device)
        
        with torch.autocast("mps"):
            outputs = model(
                pixel_values=pixel_values,
                labels=labels,
                decoder_attention_mask=decoder_attention_mask
            )
            loss = outputs.loss / grad_accum_steps
        
        loss.backward()
        running_loss += loss.item()
        
        if (step + 1) % grad_accum_steps == 0:
            optimizer.step()
            optimizer.zero_grad()
    
    print(f"Epoch {epoch+1} Loss: {running_loss/len(dataloader):.4f}")

# ---------------------------
# 6️⃣ Save trained model
# ---------------------------
output_dir = "trained_vit_gpt2_mps_fixed_length"
model.save_pretrained(output_dir)
tokenizer.save_pretrained(output_dir)
image_processor.save_pretrained(output_dir)
print(f"Model saved to {output_dir}")

# ---------------------------
# 7️⃣ Inference with max_new_tokens
# ---------------------------
model.eval()
all_generated_sequences = []

max_seq_len = 8  # match your JSON sequence length

inference_dataloader = DataLoader(dataset, batch_size=4, shuffle=False)

for batch in inference_dataloader:
    pixel_values = batch["pixel_values"].to(device)
    
    with torch.no_grad():
        generated_ids = model.generate(
            pixel_values,
            max_new_tokens=max_seq_len,        # generate only max_seq_len tokens
            eos_token_id=tokenizer.eos_token_id,
            pad_token_id=tokenizer.pad_token_id,
            num_beams=3,
            no_repeat_ngram_size=2,           # optional, prevents repetition
            early_stopping=True,
            do_sample=False
        )
    
    generated_texts = tokenizer.batch_decode(generated_ids, skip_special_tokens=True)
    # Truncate to exact token count (safety)
    generated_texts = [" ".join(t.split()[:max_seq_len]) for t in generated_texts]
    
    all_generated_sequences.extend(generated_texts)

# ---------------------------
# 8️⃣ Save generated sequences
# ---------------------------
output_json = [
    {"image": dataset.data[i]["image"], "generated_sequence": seq}
    for i, seq in enumerate(all_generated_sequences)
]

with open("generated_sequences_mps_fixed_length.json", "w") as f:
    json.dump(output_json, f, indent=4)

print("All generated sequences saved to generated_sequences_mps_fixed_length.json")
```

```python
import matplotlib.pyplot as plt
from PIL import Image
import json

# Load generated sequences JSON
with open("generated_sequences_mps_attention.json", "r") as f:
    results = json.load(f)

# Function to display a batch of images with their sequences
def visualize_results(results, image_root="data", num_samples=8):
    plt.figure(figsize=(15, num_samples * 2))
    
    for i, item in enumerate(results[:num_samples]):
        image_path = f"{image_root}/{item['image']}"
        sequence = item["generated_sequence"]
        image = Image.open(image_path).convert("RGB")
        
        plt.subplot(num_samples, 1, i + 1)
        plt.imshow(image)
        plt.axis("off")
        plt.title(sequence, fontsize=12)
    
    plt.tight_layout()
    plt.show()

# Visualize first 8 results
visualize_results(results, image_root="data", num_samples=8)

```
